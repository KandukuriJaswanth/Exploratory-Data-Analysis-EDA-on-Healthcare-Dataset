## **Data Visualization**

# Histograms for distributions.


```python
# Assuming df is already defined and contains the necessary columns
# Create a figure with subplots
plt.figure(figsize=(15, 10))

# Create histogram for Age
plt.subplot(2, 2, 1)
plt.hist(df['Age'], bins=10, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Create histogram for Billing Amount
plt.subplot(2, 2, 2)
plt.hist(df['Billing Amount'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Billing Amount Distribution')
plt.xlabel('Billing Amount')
plt.ylabel('Frequency')

# Create histogram for Room Number counts
plt.subplot(2, 2, 3)
plt.hist(df['Room Number'], bins=len(df['Room Number'].unique()), color='lightcoral', edgecolor='m')
plt.title('Distribution of Patients by Room Number')
plt.xlabel('Room Number')
plt.ylabel('Number of Patients')

# Histogram for Age distribution by Gender
plt.subplot(2, 2, 4)  # Corrected to a new subplot index
genders = df['Gender'].unique()
for gender in genders:
    plt.hist(df[df['Gender'] == gender]['Age'], bins=10, alpha=0.5, label=gender)

plt.title('Distribution of Patients by Age and Gender')
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.legend(title='Gender')

# Create a new figure for the histogram of Days Admitted by Medical Condition
plt.figure(figsize=(10, 6))
df.groupby('Medical Condition')['Days_Admitted'].hist(alpha=0.5, bins=range(1, df['Days_Admitted'].max() + 2), legend=True)
plt.title('Histogram of Days Admitted by Medical Condition')
plt.xlabel('Days Admitted')
plt.ylabel('Frequency')
plt.xticks(range(1, df['Days_Admitted'].max() + 1))
plt.legend(df['Medical Condition'].unique())
plt.grid(axis='y', alpha=0.75)



# Display the plots
plt.tight_layout()
plt.show()
```

# Scatter plots for relationships between variables

```python
plt.figure(figsize=(15, 5))
# Scatter plot for Age vs. Medical Condition
plt.subplot(1, 3, 1)
plt.scatter(df['Age'], df['Medical Condition'], color='skyblue', alpha=0.5)
plt.title('Age vs. Medical Condition')
plt.xlabel('Age')
plt.ylabel('Medical Condition')

# Normalize the age for the second scatter plot
normed_age = (df['Age'] - df['Age'].min()) / (df['Age'].max() - df['Age'].min())

# Create a scatter plot for Medical Condition vs. Age colored by normalized age
plt.subplot(1, 3, 2)
sc = plt.scatter(df['Medical Condition'], df['Age'], c=normed_age, cmap='Set2', alpha=0.7)
plt.title('Medical Condition vs. Age')
plt.xlabel('Medical Condition')
plt.ylabel('Age')

# Add color bar to represent normalized ages
cbar = plt.colorbar(sc)
cbar.set_label('Normalized Age')

# Scatter plot for Medication vs. Test Results
plt.subplot(1, 3, 3)
plt.scatter(df['Medication'], df['Test Results'], color='lightcoral', alpha=0.5)
plt.title('Medication vs. Test Results')
plt.xlabel('Medication')
plt.ylabel('Test Results')

# Adjust layout
plt.tight_layout()

# Scatter plot for Billing Amount vs. Room Number
plt.figure(figsize=(8, 5))
plt.scatter(df['Billing Amount'], df['Room Number'], color='lightgreen', alpha=0.5)
plt.title('Billing Amount vs. Room Number')
plt.xlabel('Billing Amount')
plt.ylabel('Room Number')
plt.tight_layout()
plt.show()
```

## Bar plots for categorical variables.

```python
# List of categorical variables to plot
categorical_columns = [
    'Gender', 'Blood Type', 'Medical Condition', 'Doctor',
    'Hospital', 'Insurance Provider', 'Admission Type', 'Medication', 'Test Results'
]

# Create subplots for each categorical variable
plt.figure(figsize=(15, len(categorical_columns) * 3))

for i, column in enumerate(categorical_columns, start=1):
    plt.subplot(len(categorical_columns), 1, i)
    df[column].value_counts().plot(kind='bar', color='lightblue', edgecolor='black')
    plt.title(f'{column} Distribution')
    plt.xlabel('Categories')
    plt.ylabel('Frequency')
    plt.xticks(rotation=45)  # Rotate x-axis labels for readability


plt.tight_layout()

# Show the plots
plt.show()
```
